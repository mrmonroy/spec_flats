```python
qe_dir = '/home/m/monroy/repos/obs_lsst_data/latiss/transmission_sensor/rxx_s00/'
qe_file = '19700101T000000.ecsv'
```

```python
qes = open(os.path.join(qe_dir,qe_file)).readlines()
```

```python
ampli_qes = []
for i,ampli_ in enumerate(ampli_seq):
    ampli_df_ = qes_df[qes_df['amp_name']==ampli_]
    if i==0:
        ampli_wls = np.array(ampli_df_['wavelength']*10.)
    plt.plot(ampli_wls,ampli_df_['efficiency'])
    ampli_qes.append(np.array(ampli_df_['efficiency']))
quantum_eff_ = np.average(np.array(ampli_qes),axis=0)
print(len(quantum_eff_))
plt.plot(ampli_df_['wavelength']*10.,quantum_eff_,ls='--',color='k')
```

## Load filter throughputs

```python
filter_basedir = '/home/m/monroy/repos/obs_lsst_data/latiss/transmission_filter/'
gdir = 'sdssg_65mm~empty'
zdir = 'sdssz_65mm~empty'
gfile = '20221005T000000.ecsv'
zfile = '20221005T000000.ecsv'
```

```python
thg = open(os.path.join(filter_basedir,gdir,gfile)).readlines()
thz = open(os.path.join(filter_basedir,zdir,zfile)).readlines()
```

```python
throughput_g_ = thg[29:]
throughput_z_ = thz[29:]
for i in range(len(throughput_g_)):
    throughput_g_[i] = throughput_g_[i].split()
    
    throughput_g_[i][0] = float(throughput_g_[i][0])
    throughput_g_[i][1] = float(throughput_g_[i][1])

for i in range(len(throughput_z_)):
    throughput_z_[i] = throughput_z_[i].split()
    
    throughput_z_[i][0] = float(throughput_z_[i][0])
    throughput_z_[i][1] = float(throughput_z_[i][1])
throughput_g_ = np.array(throughput_g_)
throughput_z_ = np.array(throughput_z_)
```

```python
median_blue = np.median(data_blue)
median_red = np.median(data_red)
print(median_blue,median_red)
```

```python
sigma2_ij = (median_blue/data_red)**2.+((data_blue/data_red**2.)*median_red)**2.
```

```python
_ = plt.hist(sigma2_ij.flatten(),bins=100)
plt.xscale("log")
plt.yscale("log")
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(111)
im = ax.imshow(sigma2_ij,cmap="gray",origin='lower',vmax=10)
fig.colorbar(im,ax=ax)
```

```python
def calc_chi2(k):
    chi2 = np.sum(((Rij-k*R)**2.)/(sigma2_ij))
    chi2 = chi2/len(Rij.flatten())
    return chi2
```

```python
calc_chi2(1)
```

```python
chi2 = []
k = np.linspace(-0.5,2.,50)
for k_ in k:
    chi2.append(calc_chi2(k_))
chi2 = np.array(chi2)
```

```python
plt.plot(k,chi2)
plt.axvline(x=k[chi2==np.min(chi2)],ls=':')
```

```python
chi2_min = np.min(chi2)
chi2_min
```

```python
chi2_interp = interp1d(k,chi2,bounds_error=False,fill_value=0.)
```

```python
Delta_chi2 = 1.
```

```python
from scipy.optimize import root_scalar
```

```python
k_min = k[chi2==chi2_min]
k_min
```

```python
calc_chi2(k_min),chi2_min
```

```python
def chi2_diff(k, Delta_chi2):
    return calc_chi2(k) - (chi2_min + Delta_chi2)
```

```python
k_upper = root_scalar(
    chi2_diff,
    args=(Delta_chi2,),
    bracket=[k_min, k_min + 1]  # Intervalo inicial para buscar la raz
).root
```

```python
k_lower = root_scalar(
    chi2_diff,
    args=(Delta_chi2,),
    bracket=[k_min - 1, k_min]  # Intervalo inicial para buscar la raz
).root
```

```python
k_lower,k_upper
```

```python
plt.plot(k,chi2)
plt.axvline(x=k[chi2==np.min(chi2)],ls=':',color='k')
plt.axvline(k_lower,ls=':',color='r')
plt.axvline(k_upper,ls=':',color='r')
plt.xlabel('k')
plt.ylabel(r'$\chi^2$')
```
