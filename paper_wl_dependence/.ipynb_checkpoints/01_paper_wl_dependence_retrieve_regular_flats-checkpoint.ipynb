```python
np.save(os.path.join(outdir,'LATISS_amplifier_coordinates.npy'),myflat.amplis_coords_all)
```

```python
outfile = 'regular_flat_{0}_{1}-{2}.fits'.format(myflat.flat_id,filter_,disperser)
```

```python
myflat.save('flat',outfile,outdir)
```

```python
saved_flat = fits.open(os.path.join(outdir,outfile))
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
im = ax.imshow(saved_flat[1].data,cmap="gray",origin='lower',vmin=0.98,vmax=1.02)
fig.colorbar(im,ax=ax)
```

### We can also normalize the flat 

Each amplifier is normalized by its median separately 

```python
myflat.normalize_flat()
```

```python
myflat.plot_flat(show="norm",vmin=0.97,vmax=1.03)
```

```python
outfile = 'normalised_flat_{0}_{1}-{2}.fits'.format(myflat.flat_id,filter_,disperser)
```

```python
myflat.save('norm',outfile,outdir)
```

```python
saved_flat = fits.open(os.path.join(outdir,outfile))
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
im = ax.imshow(saved_flat[1].data,cmap="gray",origin='lower',vmin=0.98,vmax=1.02)
fig.colorbar(im,ax=ax)
```
