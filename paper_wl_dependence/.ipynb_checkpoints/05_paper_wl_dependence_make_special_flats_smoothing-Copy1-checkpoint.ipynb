### Tests on window size 

```python
ws = list(np.linspace(10,19,10).astype('int'))#+[20,30,40,50]
ws
```

```python
myflat.amplis_coords
```

```python
yref1 = 3250
yref2 = 2500
xref1 = 1338
xref2 = 1268
```

```python

smooth_arrays_x = {}
smooth_arrays_y = {}
for ws_ in ws:
    myflat.smooth_flat(window_size=ws_,sep_amplis='together',normalize=False)
    smooth_arrays_x[ws_] = [myflat.smooth_array[yref1,xrange_],myflat.smooth_array[yref2,xrange_]]
    smooth_arrays_y[ws_] = [myflat.smooth_array[yrange1_,xref1],myflat.smooth_array[yrange2_,xref2]]
    
```

### Comparison of window sizes with normalised flat

### Horizontal analysis 

```python
sel_window = 15
```

```python

colors = plt.cm.cividis(np.linspace(0, 1, len(ws)))

fig = plt.figure(figsize=(7,9))
ax1 = fig.add_subplot(211)

ax1.plot(xrange_,myflat.flat_array[yref1,xrange_],ls='--',alpha=0.4,label='Norm. flat')

for ws_ in ws:
    ax1.plot(xrange_,smooth_arrays_x[ws_][0],label=ws_)

ax1.grid()

ax1.set_xlabel('X axis')
ax1.set_ylabel('ADU')

ax1.legend(loc="upper right")

x_min_zoom = 1200 
x_max_zoom = 1300 

mask = (xrange_ >= x_min_zoom) & (xrange_ <= x_max_zoom)
x_zoom = xrange_[mask]
print(len(x_zoom))

ax2 = fig.add_subplot(212)
ax2.plot(x_zoom, myflat.flat_array[yref1, x_zoom], ls='--', alpha=0.4)

for ws_ in ws:
    ax2.plot(x_zoom,smooth_arrays_x[ws_][0][mask],label=ws_)

ax2.grid()

ax2.set_xlabel('X axis')
ax2.set_ylabel('ADU')


ax1.axvspan(x_min_zoom, x_max_zoom, color='gray', alpha=0.2, label='Regin zoom')

plt.savefig(os.path.join(outdir,'horizontal_analysis.png'))
```

```python
ampli_median_ = np.median(myflat.flat_array)
```

```python
# Paleta de colores para las lneas
colors = plt.cm.cividis(np.linspace(0, 1, len(ws)))

fig = plt.figure(figsize=(7,9))
ax1 = fig.add_subplot(211)

ax1.plot(xrange_,myflat.flat_array[yref1,xrange_],ls='--',label='Flat')
ax1.plot(xrange_,myflat.flat_array[yref1,xrange_]/smooth_arrays_x[sel_window][0],ls='-',alpha=0.7,label='Flat/Smoothed')
ax1.axhline(y=ampli_median_,ls=':',color='k',label='Ampli. median')

ax1.grid()

ax1.legend()

x_min_zoom = 1200 
x_max_zoom = 1300

mask = (xrange_ >= x_min_zoom) & (xrange_ <= x_max_zoom)
x_zoom = xrange_[mask]
print(len(x_zoom))

ax2 = fig.add_subplot(212)
ax2.plot(x_zoom, myflat.flat_array[yref1, x_zoom], ls='--', alpha=0.8)
ax2.plot(x_zoom, myflat.flat_array[yref1, x_zoom]/smooth_arrays_x[sel_window][0][mask], ls='--', alpha=0.8)
ax2.axhline(y=ampli_median_,ls=':',color='k')

ax2.grid()

ax2.set_xlabel('X axis')
ax2.set_ylabel('ADU')


ax1.axvspan(x_min_zoom, x_max_zoom, color='gray', alpha=0.2, label='Regin zoom')

plt.savefig(os.path.join(outdir,'horizontal_detrending.png'))

```

```python
def plane(xx,yy,coeffs):
    a,b,c = coeffs
    return a*xx+b*yy+c
```
