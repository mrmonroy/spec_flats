```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
x0 = np.random.uniform(-1.5,1.5,50)
m = 0.2
```

```python
x = np.linspace(-2,2,100)
```

```python
xdust = np.linspace(-0.2,0.2,100)
ydust = np.ones(len(xdust))*0.2
```

```python
def fun(x,x0,m):
    return m*(x-x0)
```

```python
# Crear figura 3D con fondo oscuro
fig = plt.figure(figsize=(4, 7), facecolor='white')
ax1 = fig.add_subplot(211,facecolor='black')
light = []
for i,x0_ in enumerate(x0):
    #if i==49:
    if np.isclose(fun(xdust,x0_,m),ydust,atol=1e-2).any() or np.isclose(fun(xdust,x0_,-m),ydust,atol=1e-2).any():
        color = 'r'
        alpha = 1.0
        light.append(0.5)
    else:
        color = 'b'
        alpha = 0.5
        light.append(1.0)
        
    ax1.plot(x,fun(x,x0_,m),color=color,alpha=alpha)
    ax1.plot(x,fun(x,x0_,-m),color=color,alpha=alpha)
    
ax1.plot(xdust,ydust,linewidth=4.,color='grey')

ax2 = fig.add_subplot(212,facecolor='black')
for x0_,l_ in zip(x0,light):
    if l_==1.0:
        color='b'
    else:
        color='r'
    ax2.scatter(x0_,l_,ls='',marker='o',color=color)

ax1.set_xlim(-2,2)
ax1.set_ylim(0.,0.25)
ax2.set_xlim(-2,2)
```
