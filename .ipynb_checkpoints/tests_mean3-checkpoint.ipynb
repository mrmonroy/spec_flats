### We can also create a smoothed flat 

This is independent of the normalizing step, that is, we can do it even if we haven't done myflat.normalize_flat(), it will be done internally by default (normalize = True, very fast operation) 

```python
myflat.special_flat(window_size=40,mode='mirror')
```

```python
%matplotlib inline
myflat.plot_flat(show="special",vmin=0.97,vmax=1.03)
```

```python
myflat.smooth_array
```

```python
flat_array = myflat.flat_array[y0_:y1_,x0_:x1_]
norm_array = myflat.norm_array[y0_:y1_,x0_:x1_]
inter_array = myflat.inter_array[y0_:y1_,x0_:x1_]
smooth_array = myflat.smooth_array[y0_:y1_,x0_:x1_]
special_array = myflat.special_array[y0_:y1_,x0_:x1_]
```

```python
(special_array==flat_array/smooth_array).all()
```

```python
%matplotlib widget
fig, (ax_orig, ax_norm, ax_inter, ax_smooth, ax_special) = plt.subplots(1, 5, figsize=(8, 4))

ax_orig.imshow(flat_array, cmap='gray',origin='lower',vmin=0.97,vmax=1.03)
ax_orig.set_title('Original')
ax_orig.set_axis_off()

ax_norm.imshow(norm_array, cmap='gray',origin='lower',vmin=0.97,vmax=1.03)
ax_norm.set_title('Normalized')
ax_norm.set_axis_off()

ax_inter.imshow(inter_array, cmap='gray',origin='lower',vmin=0.97,vmax=1.03)
ax_inter.set_title('Intermediate')
ax_inter.set_axis_off()

ax_smooth.imshow(smooth_array, cmap='gray',origin='lower',vmin=0.97,vmax=1.03)
ax_smooth.set_title('Smooth')
ax_smooth.set_axis_off()

ax_special.imshow(special_array, cmap='gray',origin='lower',vmin=0.97,vmax=1.03)
ax_special.set_title('Special')
ax_special.set_axis_off()

#fig.show()
```

```python
norm_array2 = flat_array/np.median(flat_array)
```

```python
smooth_array2 = np.ones(flat_array.shape)
smooth_array2[:,:] = norm_array2[:,:]
print((smooth_array2==norm_array2).all())
```

```python
mask2 = (smooth_array2>=np.percentile(smooth_array2.flatten(),1.))*(smooth_array2<=np.percentile(smooth_array2.flatten(),99.))
```

```python
masked_array2 = np.ma.array(smooth_array2,mask=mask2,fill_value=-999)
```

```python
inter_array2 = np.copy(smooth_array2)
```

```python
inter_array2[~masked_array2.mask] = np.median(smooth_array2.flatten())
```

```python
np.median(smooth_array2.flatten())
```
